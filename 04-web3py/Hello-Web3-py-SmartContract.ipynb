```python
import json
from pprint import pprint

import web3

#Web3 Entery point
from web3 import Web3 

#Actual and Testing Provider
from web3 import HTTPProvider

#Compiler 
from solc import compile_source
from web3.contract import ConciseContract


w3 = Web3(HTTPProvider("http://127.0.0.1:8546"))
#w3 = Web3(TestRPCProvider(host='127.0.0.1', port=7545,))
```

```python
import time
def mine3alaKhafeif():
    print("Blocks Before:{}".format(w3.eth.blockNumber))
    w3.miner.start(8)
    time.sleep(1)
    w3.miner.stop()
    print("Blocks After:{}".format(w3.eth.blockNumber))

def getAllBalances():
    accounts = w3.eth.accounts
    for account in accounts:
        balance = w3.eth.getBalance(account)
        print("{} : {}".format(account, balance))
    
accounts = w3.eth.accounts
getAllBalances()    
```

## 1-Create contract

- Edit in remix and then copy here https://remix.ethereum.org

```python
# Solidity source code
contract_source_code = '''
pragma solidity ^0.4.0;

contract Greeter {
    string public greeting;

    function Greeter() {
        greeting = 'Hello';
    }

    function setGreeting(string _greeting) public payable{
        greeting = _greeting;
    }

    function greet() constant returns (string) {
        return greeting;
    }
    
    function () public payable {}
}
'''
```

```python
compiled_sol = compile_source(contract_source_code) # Compiled source code
pprint(compiled_sol)
```

```python
abi= compiled_sol['<stdin>:Greeter']['abi']
pprint(abi)
```

```python
asm= compiled_sol['<stdin>:Greeter']['asm']
pprint(asm)
```

```python
# Instantiate and deploy contract
contract_interface = compiled_sol['<stdin>:Greeter']
contract = w3.eth.contract(abi=contract_interface['abi'], bytecode=contract_interface['bin'])
contract
```

# 2- Deploy Contract

```python
# Get transaction hash from deployed contract
w3.personal.unlockAccount(accounts[0], 'foobar')
tx_hash = contract.deploy(transaction={'from': w3.eth.accounts[0], 'gas': 410000})
tx_hash
```

```python
mine3alaKhafeif()
```

```python
# Get tx receipt to get contract address
tx_receipt = w3.eth.getTransactionReceipt(tx_hash)
tx_receipt
```

```python
contract_address = tx_receipt['contractAddress']
contract_address
```

# 3- Interact with Contract

```python
# Contract instance in concise mode
abi = contract_interface['abi']
contract_instance = w3.eth.contract(address=contract_address, abi=abi,ContractFactoryClass=ConciseContract)

#Getters + Setters for web3.eth.contract object
print('Contract value: {}'.format(contract_instance.greet()))

w3.personal.unlockAccount(accounts[0], 'foobar')
contract_instance.setGreeting('Salamo 3aleko', transact={'from': w3.eth.accounts[0], 'value':5000, 'gas':41000})
```

```python
print('Contract value: {}'.format(contract_instance.greet()))
```
